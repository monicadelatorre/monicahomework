```python
url2015='https://www.usclimatedata.com/climate/los-angeles/california/united-states/usca1339/2015/1'
```

```python
#read first table
df = pd.read_html(url2015)[0]
#read second table
df1 = pd.read_html(url2015)[1]
#convert tables to pandas dataframes
table_df=pd.DataFrame(df)
table_df1=pd.DataFrame(df1)
```

```python
#show first table
table_df
```

```python
#show second table, remove index
table_df1=table_df1.drop(table_df1.index[[0]],axis=1)
table_df1
```

```python
#concatenate both tables horizontally and remove row 4 and 5
result = pd.concat([table_df, table_df1], axis=1, sort=False)
result = result.drop(result.index[[4,5]])
result
```

```python
result = result.rename_axis(None)
```

```python
result
```

```python
result['average'] = result.mean(numeric_only=True, axis=1)
result
```
